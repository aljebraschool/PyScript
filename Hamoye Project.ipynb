### Use the New dataframe to build Correlation (Select the most correlating features)

```python

# Calculate the correlation matrix
corr_matrix = df.corr()

# Sort the correlations with 'Total' in descending order
corr_with_output = corr_matrix['Total'].abs().sort_values(ascending=False)

# Set your correlation threshold here
threshold = 0.1  

# Select relevant features based on the threshold
relevant_features = corr_with_output[corr_with_output > threshold]

# Plot the correlation values
plt.figure(figsize=(10, 8))
relevant_features.plot(kind='bar', color='skyblue')
plt.title('Correlation with Total')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.xticks(rotation=45)
plt.show()
```

```python
#remove the less correlating, target column to create you input feature and convert dataframe to numpy using (values)
X = df.drop(columns=['Latitude', 'Longitude', 'UN Member', 'Total']).values

#visualize your features
print('X: ', X)
print('X shape: ', X.shape)
```

```python
#create your output variable in numpy
y = df['Total'].values

#visualize your target variable
print('y:', y)
print('y shape:', y.shape)
```

# Model Building 

```python
#import needed libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#split your dataset into train, test using test size of 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

```

```python
#import library to normalize your dataset
from sklearn.preprocessing import StandardScaler

#create class object 
scaler = StandardScaler()

#normalize your data
X_train = scaler.fit_transform(X_train) #use a defined mean with (fit)
X_test = scaler.transform(X_test) #use the same mean without (fit)

print(X_train)
```

### Build Linear Regressiong

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#make predictions with the test data
y_pred = model.predict(X_test)

#compare your predicted values with real value
np.set_printoptions(precision=2)
print('Predicted       Actual')

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

#see error in predictions
print("Error : ", mean_squared_error(y_test, y_pred))
     
```

### Building RandomForestRegressor 

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model.fit(X_train, y_train)
```

```python
#make predictions with the test data
y_pred = model.predict(X_test)

#compare your predicted values with real value
np.set_printoptions(precision=2)
print('Predicted       Actual')

print(np.concatenate(   (y_pred.reshape(-1,1), y_test.reshape(-1,1)),1))

#see error in predictions
print("Error : ", mean_squared_error(y_test, y_pred))
```

### Building Support Vector Regression

```python
from sklearn.svm import SVR

model = SVR()

model.fit(X_train, y_train)
```

```python
#make predictions with the test data
y_pred = model.predict(X_test)

#compare your predicted values with real value
np.set_printoptions(precision=2)
print('Predicted       Actual')

print(np.concatenate(   (y_pred.reshape(-1,1), y_test.reshape(-1,1)),1))

#see error in predictions
print("Error : ", mean_squared_error(y_test, y_pred))
```
